### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",sep="|")
users.head(5)
```

### Step 4. Discover what is the mean age per occupation

```python
df = pd.DataFrame(users)
mean_age_per_occupation = df.groupby("occupation")["age"].mean()
mean_age_per_occupation
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
# 각 직업별 남성 비율(Male ratio)을 계산하고, 이를 내림차순으로 정렬하는 문
occupation_counts = df.groupby('occupation')['gender'].count() # 직업별 총 인원 수
male_counts = df[df['gender'] == 'M'].groupby('occupation')['gender'].count() # 직업별 남성 인원 수
male_ratio = male_counts / occupation_counts # 직업별 남성 비율
male_ratio.sort_values(ascending=False) # 직업별 남성 비율을 내림차순으로 정렬
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
age_stats_per_occupation = df.groupby('occupation')['age'].agg(['min', 'max'])
age_stats_per_occupation
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
#각 직업(occupation)과 성별(gender) 조합에 대해 평균 나이(age)를 계산
mean_age = df.groupby(['occupation', 'gender'])['age'].mean().reset_index()
mean_age
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_counts = df.groupby(['occupation', 'gender']).size().unstack(fill_value=0)
total_counts = gender_counts.sum(axis=1)
gender_percentages = gender_counts.div(total_counts, axis=0) * 100
gender_percentages
```
