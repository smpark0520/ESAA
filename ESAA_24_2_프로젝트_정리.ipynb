1. 연면적 적정성

연면적 : 건물 전체 면적, 건물의 각 층 면적을 모두 합한 총면적

분양면적 : 실제 분양되는 면적, 전용면적+공용면적

(성남시 사례 분석 결과)
- 피처중요도 가장 높음, SHAP 결과에서도 공실률에 높은 영향력 확인
- 연면적이 큰 관측치들은 SHAP에서 오른쪽에 위치 = 연면적↑ 공실률↑ 경향

-> 연면적 축소, 초기 분양면적 조정 등을 통해 수요와 공급의 균형 유지
