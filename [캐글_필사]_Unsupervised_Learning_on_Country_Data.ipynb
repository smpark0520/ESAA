# 데이터셋 소개

- HELP International을 위한 비지도 학습을 이용한 국가 군집화
- 목표: 국가의 전반적인 발전 수준을 결정짓는 사회경제적 및 건강 요인을 사용하여 국가들을 범주화하는 것.

- 조직 소개 : HELP International은 빈곤 퇴치 및 재난 및 자연재해 시 기본적인 생활 필수품과 구호를 제공하는 데 헌신하는 국제 인도주의 비정부 기구(NGO)

- 문제 정의 : HELP International은 약 1천만 달러의 자금을 조달함. 이제 NGO의 CEO는 이 자금을 전략적이고 효과적으로 사용할 방법을 결정해야 함. 따라서 CEO는 가장 절실히 지원이 필요한 국가를 선택해야 함.

-> 사회경제적 및 건강 요인을 기반으로 국가를 군집화한 뒤, CEO가 우선적으로 집중해야 할 국가를 제안하는 것

데이터셋 변수 설명

1. country (문자열)

- 국가 이름.
고유 식별자로 사용되며, 분석에서 제외하거나 결과에 포함할 수 있음.

2. child_mort (숫자)

- 5세 이하 어린이의 사망률 (출생 1000명당 사망 수).
값이 높을수록 해당 국가의 유아 건강 상황이 열악함을 나타냄.

3. exports (숫자, %)

- 1인당 상품 및 서비스 수출 비율 (GDP 대비 백분율).
경제의 국제 교역 의존도를 나타냄.

4. health (숫자, %)

- 1인당 건강 관련 지출 (GDP 대비 백분율).
값이 높을수록 해당 국가가 건강에 많은 투자를 하고 있음을 의미.

5. imports (숫자, %)

- 1인당 상품 및 서비스 수입 비율 (GDP 대비 백분율).
국제 경제 상호작용을 나타냄.

6. income (숫자)

- 1인당 순소득 (Net Income per Person).
개인의 평균 소득 수준을 나타냄.

7. inflation (숫자, %)

- 연간 GDP 성장률 (인플레이션 측정치).
값이 높으면 경제 불안정 가능성을 시사할 수 있음.

8. life_expec (숫자)

- 평균 기대 수명 (현재의 사망률 패턴이 유지된다고 가정할 때 신생아의 예상 수명).
값이 높을수록 국가의 보건 및 삶의 질 수준이 높음을 의미.

9. total_fer (숫자)

- 평균 여성 1명당 출산율 (현재의 연령별 출산율이 유지된다고 가정할 때).
높은 값은 인구 증가율이 빠른 국가를 나타냄.

10. gdpp (숫자)

- 1인당 GDP (총 GDP를 전체 인구로 나눈 값).
국가의 경제적 발전 수준을 나타냄.

# PCA

```python
pca = PCA(svd_solver='randomized', random_state=42, n_components=9).fit(df)
exp = pca.explained_variance_ratio_
exp
```

- 첫 번째 주성분: 0.4595 (45.95%)
: 첫 번째 주성분이 데이터에서 가장 많은 분산(45.95%)을 설명

- 두 번째 주성분: 0.1718 (17.18%)
: 두 번째 주성분이 데이터에서 17.18%의 분산을 설명

- 세 번째 주성분: 0.1300 (13.00%)
: 세 번째 주성분은 13.00%의 분산을 설명

- 네 번째 주성분: 0.1105 (11.05%)
: 네 번째 주성분은 11.05%의 분산을 설명

- 다섯 번째 주성분까지 포함하면 94.53%를 설명

```python
plt.plot(np.cumsum(exp), linewidth=3)
plt.title('n_component - Cumulative Explained Variance')
plt.xlabel('Number of n_components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.xticks(range(0, 10, 1))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.show()
```

```python
n=1
for component in pca.components_:
  print('n_component=',n)
  print(component)
  print('_'*50)
  n+=1
```

- 주성분 분석(PCA)에서 각 주성분의 벡터를 출력하는 코드
- 각 주성분은 데이터의 각 특성에 대한 가중치를 포함하는 벡터로, 이를 통해 각 주성분이 원본 데이터의 특성들에 어떤 영향을 미치는지 확인 가능

```python
# We use n_component=5
pca_df = pd.DataFrame({
    'Features' : columns[1:],
    'PC1' : pca.components_[0],
    'PC2' : pca.components_[1],
    'PC3' : pca.components_[2],
    'PC4' : pca.components_[3],
    'PC5' : pca.components_[4],
})
pca_df
```

```python
plt.scatter(pca_df.PC1, pca_df.PC2)
plt.xlabel('PC1')
plt.ylabel('PC2')
for i, txt in enumerate(pca_df.Features):
  plt.annotate(txt, (pca_df.PC1[i], pca_df.PC2[i]))
plt.show()
```

# Clustering

## KMeans

```python
kmeans_set = {
    'init' : 'random',
    'n_init' : 10,
    'max_iter' : 300,
    'random_state' : 42
}

list_kmeans = []

for k in range(2,10):
  kmeans = KMeans(n_clusters=k, **kmeans_set)
  kmeans.fit(pcs_df2)
  list_kmeans.append(kmeans.inertia_)
```

- 클러스터 개수를 2부터 9까지 변화시키면서 각 클러스터 개수에 따른 관성값 (inertia: 각 데이터 포인트가 가장 가까운 클러스터 중심과 가지는 거리의 제곱합)을 계산

```python
list_kmeans
```

```python
print('Minimum inertia', min(list_kmeans))
print('Number of clusters :', list_kmeans.index(min(list_kmeans))+2)
print('_'*50)
print('Maximum inertia', max(list_kmeans))
print('Number of clusters :', list_kmeans.index(max(list_kmeans))+2)
```

<최소 관성값 (Minimum Inertia)>

k=9, 356.39897763197314

→ 클러스터 개수가 많아질수록 관성값이 작아지는데, 이는 클러스터가 세분화되며 각 데이터 포인트가 더 가까운 클러스터 중심에 배정되기 때문

<최대 관성값 (Maximum Inertia)>

k=2, 968.1808440726793


→ 클러스터가 2개일 때 데이터의 분산(관성값)이 가장 큼. 이는 데이터가 두 개의 클러스터로 나뉘는 과정에서 큰 변화를 겪기 때문

```python
k1 = KneeLocator(range(2,10),list_kmeans,curve='convex',direction='decreasing')
k1.elbow
```

- 클러스터 개수를 4로 설정하는 것이 데이터의 분산(관성값)과 모델의 복잡성 간의 균형을 가장 잘 맞춘다는 것

```python
plt.plot(range(2,10),list_kmeans,linewidth=3,label='Inertia')
plt.axvline(x=k1.elbow,ls='--',label='elbow')
plt.axvline(x=list_kmeans.index(max(list_kmeans))+2, alpha=0.3,ls='--',label='maximum of inertia',color='green')
plt.axvline(x=list_kmeans.index(min(list_kmeans))+2, alpha=0.3,ls='--',label='maximum of inertia',color='red')
plt.axvline(x=3,ls='--',label='i think this is best',color='purple')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.legend()
plt.show()
```

## Silhouette

```python
# Calculate silhouette score for kmeans
kmeans_sil_score=0
kmeans_sil_n=0
silhouette_coefficient = []
for k in range(2,10):
  kmeans = KMeans(n_clusters=k, **kmeans_set).fit(pcs_df2)
  score = silhouette_score(pcs_df2,kmeans.labels_)
  silhouette_coefficient.append(score)
  if score > kmeans_sil_score:
    kmeans_sil_score = score
    kmeans_sil_n = k
  else:
    continue
print('Best Silhouette_score:', kmeans_sil_score)
print('Best n_cluster:', kmeans_sil_n)

plt.plot(range(2, 10), silhouette_coefficient, linewidth=3)
plt.xticks(range(2,10))
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Coefficient')
plt.show()
```

## calinski_harabasz_score

```python
# Calculate calinski_harabasz_score for kmeans
kmeans_ch_score=0
kmeans_ch_n=0
calinski_coefficient=[]
for k in range(2,10):
  kmeans=KMeans(n_clusters=k, **kmeans_set).fit(pcs_df2)
  score=calinski_harabasz_score(pcs_df2, kmeans.labels_)
  calinski_coefficient.append(score)
  if score > kmeans_ch_score:
    kmeans_ch_score = score
    kmeans_ch_n = k
  else:
    continue
print('Best calinski_harabasz_score:', kmeans_ch_score)
print('Best n_cluster : ', kmeans_ch_n)

plt.plot(range(2, 10), calinski_coefficient, linewidth=3)
plt.xticks(range(2, 10))
plt.xlabel('Number Of Clusters')
plt.ylabel('calinski_harabasz_score')
plt.show()
```

# DBSCAN

```python
list_sil = []
list_davies = []
list_ch = []
list_n = []
list_noise = []
list_eps = []

for i in range(0, 40, 1):
    e = 0.5 + (i*0.05)
    dbscan = DBSCAN(eps=e)
    dbscan.fit(pcs_df2)

    labels = dbscan.labels_
    if len(labels) > 1 :

        # Number of clusters in labels, ignoring noise if present.
        n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise_ = list(labels).count(-1)

        if n_clusters_ > 1 :

            list_eps.append(e)
            list_n.append(n_clusters_)
            list_noise.append(n_noise_)
            sil = silhouette_score(pcs_df2, labels)
            davies = davies_bouldin_score(pcs_df2, labels)
            ch = calinski_harabasz_score(pcs_df2, labels)

            list_sil.append(sil)
            list_davies.append(davies)
            list_ch.append(ch)

            print("eps = ", e)
            print("Estimated number of clusters: %d" % n_clusters_)
            print("Estimated number of noise points: %d" % n_noise_)
            print('Silhouette : ', sil)
            print('Davis : ', davies)
            print('CH Score : ',   ch)
            print('\n____________________________________________________________________\n')
print("END")
```
