### Step 1. Import the necessary libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

### Step 2. Create the DataFrame that should look like the one below.

raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],

````python
raw_data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'female': [0, 1, 1, 0, 1],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(raw_data)

print(df)
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize=(10, 6))
plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age']*10, alpha=0.5)


plt.title('Scatterplot of preTestScore vs postTestScore', fontsize=15)
plt.xlabel('preTestScore', fontsize=12)
plt.ylabel('postTestScore', fontsize=12)

plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
plt.figure(figsize=(10, 6))

# seaborn의 scatterplot을 사용하여 성별에 따라 색상 구분
sns.scatterplot(
    x='preTestScore',
    y='postTestScore',
    size=df['postTestScore'] * 4.5,
    hue='female',
    palette={0: 'blue', 1: 'pink'},
    data=df,
    alpha=0.6
)

plt.title('Scatterplot of preTestScore vs postTestScore', fontsize=15)
plt.xlabel('preTestScore', fontsize=12)
plt.ylabel('postTestScore', fontsize=12)

plt.show()
````

### BONUS: Create your own question and answer it.

Scatterplot을 그릴 때, 각 점의 크기가 나이(age)와 postTestScore의 곱으로 결정되고, 점의 색상은 preTestScore의 범위(낮음, 중간, 높음)에 따라 다르게 표현되도록 해보세요.

````python
# preTestScore 구간 나누기 (낮음, 중간, 높음)
bins = [0, 10, 20, 100]
labels = ['Low', 'Medium', 'High']
df['preTestScore_category'] = pd.cut(df['preTestScore'], bins=bins, labels=labels)

plt.figure(figsize=(10, 6))

sns.scatterplot(
    x='preTestScore',
    y='postTestScore',
    size=df['age'] * df['postTestScore'],
    hue='preTestScore_category',  # preTestScore 범주에 따라 색상 구분
    palette={'Low': 'green', 'Medium': 'orange', 'High': 'red'},
    data=df,
    alpha=0.6
)

plt.title('Scatterplot of preTestScore vs postTestScore (Size by Age * postTestScore)', fontsize=15)
plt.xlabel('preTestScore', fontsize=12)
plt.ylabel('postTestScore', fontsize=12)

plt.show()
````
