## 1.2.2 머신 러닝 학습 알고리즘


### 1. 지도 학습 (Supervised Learning)
- **정의**: 입력 데이터와 함께 정답(레이블)을 제공하여 학습시키는 방식
- **예시**: "7 x 4 = ?"라는 문제와 함께 정답 28을 알려주는 것
- **특징**: 명확한 정답이 존재하며 분류(Classification)와 회귀(Regression)에 사용됨

### 2. 비지도 학습 (Unsupervised Learning)
- **정의**: 정답 없이 데이터만을 이용해 패턴이나 구조를 학습하는 방식
- **예시**: 비슷한 동물들(토끼, 다람쥐 등)을 클러스터링하여 그룹화
- **특징**: 데이터 간 유사도를 기반으로 클러스터링 수행, 명확한 정답 없음

### 3. 강화 학습 (Reinforcement Learning)
- **정의**: 에이전트가 환경과 상호작용하며 보상을 통해 최적의 행동을 학습하는 방식
- **예시**: 게임(예: 쿠키런)에서 캐릭터가 보상을 얻기 위해 행동을 선택함
- **특징**: 정답이 없고, 보상을 최대화하는 방향으로 스스로 학습함



# 1.3 딥러닝이란

- 인간의 신경망 원리를 모방한 심층 신경망 이론을 기반으로 하는 머신 러닝의 한 종류
- 기존 머신 러닝과의 큰 차이점은 인간의 뇌 구조(뉴런과 시냅스)를 기초로 하여 설계했다는 점

### 딥러닝의 핵심 구조: 인공 신경망
- 뉴런(Neuron): 인간의 뇌처럼 정보를 처리하는 기본 단위
- 시냅스(Synapse): 뉴런과 뉴런 사이의 연결 부위

-> 이 구조를 컴퓨터에 적용하여 병렬 연산을 통해 음성 인식, 이미지 처리 등 기존에 어려웠던 문제 해결이 가능해짐





### 🧠 인간 신경망 구성 요소 (딥러닝에서 모사한 구조)

| 구성 요소    | 설명 |
|--------------|------|
| **수상돌기**   | 다른 뉴런이나 주변에서 자극을 받아들이고 전기 신호로 세포체와 축삭돌기로 전달 |
| **시냅스**     | 한 뉴런의 축삭돌기와 다음 뉴런의 수상돌기가 만나는 연결 부위 |
| **축삭돌기**   | 전기 신호를 다음 뉴런으로 전달하는 긴 돌기 |
| **축삭말단**   | 전기 신호를 받아 신경 전달 물질을 시냅스 틈새로 방출 |

## 2.2.1 텐서 다루기

### 텐서 생성 및 변환

- **텐서 생성**:
  - 2차원 텐서 생성: `torch.tensor([[1, 2], [3, 4]])`
  - GPU에 텐서 생성: `torch.tensor([[1, 2], [3, 4]], device="cuda:0")`
  - 데이터 타입 지정하여 텐서 생성: `torch.tensor([[1, 2], [3, 4]], dtype=torch.float64)`

- **텐서 변환**:
  - 텐서를 ndarray로 변환: `temp.numpy()`
  - GPU 텐서를 CPU로 변환 후 ndarray로 변환: `temp.to("cpu").numpy()`


````python
import torch
print(torch.tensor([[1,2], [3,4]])) # 2차원 형태의 텐서 생성
print(torch.tensor([[1,2], [3,4]])) # GPU에 텐서 생성
print(torch.tensor([[1,2], [3,4]], dtype=torch.float64)) # dtype을 이용하여 텐서 생성
````

- 텐서를 ndarray로 변환

````python
temp = torch.tensor([[1,2], [3,4]])
print(temp.numpy()) # 텐서를 ndarray로 변환

temp = torch.tensor([[1,2], [3,4]])
print(temp.to('cpu').numpy()) # GPU 상의 텐서를 CPU의 텐서로 변환한 후 ndarray로 변환
````

## 2.2.3 모델 정의

계층 (Layer)
- 하나의 연산 단위 (예: 선형 계층, 합성곱 계층 등)
- 예: nn.Linear, nn.Conv2d 등

모듈 (Module)
- 계층 하나 또는 여러 계층을 포함하는 단위
- 모든 계층과 모델은 nn.Module을 상속받아 만듦
- 모듈 안에 또 다른 모듈을 포함시켜 계층적 구성이 가능

모델 (Model)
- 하나의 완성된 신경망 구조
- 일반적으로 하나의 큰 nn.Module 클래스 형태
- 최종 학습과 추론 대상

### 단순 신경망을 정의하는 방법

model = nn.Linear(in_features=1, out_features=1, bias=True)

### nn.Module( )을 상속하여 정의하는 방법
- __init__() : 사용할 계층(예: nn.Linear, nn.Conv2d)과 활성화 함수 등을 정의

- forward() : 실제 연산 수행 정의 (입력 → 계층 → 활성화 등)



- 파이토치에서 모델을 정의하는 코드

````python
from torch.nn import Module, Linear, Sigmoid

class MLP(Module):
  def __init__(self, inputs):
    super(MLP, self).__init__()
    self.layer = Linear(inputs, 1) # 계층 정의
    self.activation = Sigmoid() # 활성화 함수 정의

  def forward(self, X):
    X = self.layer(X)
    X = self.activation(X)
    return X
````

### Sequential 신경망 정의하는 방법

- nn.Sequential은 여러 계층(Module)을 순차적으로 연결하여 모델을 구성함

- __init__()에서 계층만 정의하고, forward()에서는 명시적으로 연산을 하지 않아도 됨

- 가독성 좋고 코드가 간결함

````python
import torch.nn as nn
class MLP(nn.Module):
  def __init__(self):
    super(MLP, self).__init__()
    self.layer1 = nn.Sequential(
        nn.Conv2d(in_channels=3, out_channels=64, kernel_size=5),
        nn.ReLU(inplace=True),
        nn.MaxPool2d(2))

    self.layer2 = nn.Sequential(
        nn.Conv2d(in_channels=64, out_channels=30, kernel_size=5),
        nn.ReLU(inplace=True),
        nn.MaxPool2d(2))

    self.layer3 = nn.Sequential(
        nn.Linear(in_features=30*5*5, out_features=10, bias=True),
        nn.ReLU(inplace=True))

    def forward(self, x):
      x = self.layer1(x)
      x = self.layer2(x)
      x = x.view(x.shape[0], -1)
      x = self.layer3(x)
      return x
model = MLP() # 모델에 대한 객체 생성

print("Printing Children\n---------------------------")
print(list(model.children()))
print("\n\nPrinting Modules\n---------------------------")
print(list(model.modules()))
````

- 함수 정의: 계층 재사용 가능  
- 단점: 모델 복잡 시 관리 어려움  
- 복잡한 모델: nn.Module 상속 권장

````python
def MLP(in_features=1, hidden_features=20, out_features=1):
  hidden = nn.Linear(in_features=in_features, out_features=hidden_features,
                     bias=True)
  activation = nn.ReLU()
  output = nn.Linear(in_features=hidden_features, out_features=out_features,
                     bias=True)
  net = nn.Sequential(hidden, activation, output)
  return net
````

## 2.2.4 모델의 파라미터 정의



#### 손실 함수 (Loss Function)
- **BCELoss**: 이진 분류
- **CrossEntropyLoss**: 다중 클래스 분류
- **MSELoss**: 회귀 모델

#### 옵티마이저 (Optimizer)
- 파라미터 업데이트를 담당
- 주요 메서드:
  - `step()`: 파라미터 업데이트
  - `zero_grad()`: 기울기 초기화
- 대표 옵티마이저:
  - `optim.Adam`, `optim.SGD`, `optim.RMSprop`, `optim.Adagrad`, `optim.Adamax` 등

#### 학습률 스케줄러 (Learning Rate Scheduler)
- 에포크 수에 따라 학습률 조정
- 주요 스케줄러:
  - `LambdaLR`: 람다 함수 기반
  - `StepLR`: 일정 스텝마다 감소
  - `MultiStepLR`: 지정된 에포크에서 감소
  - `ExponentialLR`: 매 에포크마다 감소
  - `CosineAnnealingLR`: 코사인 함수 형태
  - `ReduceLROnPlateau`: 성능 정체 시 감소

#### 지표 (Metrics)
- 학습 및 테스트 중 성능 모니터링용


####  전역 최소점과 최적점

- **손실 함수**: 실제 값과 예측 값의 차이를 수치화하는 함수
  - 오차 값이 클수록 손실 함수 값이 크고, 오차 값이 작을수록 손실 함수 값이 작음
  - 학습의 목표는 손실 함수를 최소화하는 가중치와 바이어스를 찾는 것

- **전역 최소점 (Global Minimum)**:
  - 오차가 가장 작을 때의 값으로, 우리가 최종적으로 찾고자 하는 최적점

- **지역 최소점 (Local Minimum)**:
  - 전역 최소점을 찾는 과정에서 만나는 국소적인 최솟값
  - 옵티마이저가 지역 최소점에서 멈추면 최종적인 최솟값을 찾지 못하는 문제 발생

## 2.2.5 모델 훈련


1. **기울기 초기화**:
   - `optimizer.zero_grad()` 메서드를 사용하여 기울기를 초기화
   - 기울기는 `loss.backward()`에서 계산되며, 이전 기울기 값에 누적됨. 따라서 매 학습마다 기울기를 초기화해야 함

2. **파라미터 수정**:
   - 모델의 파라미터(w, b)는 `loss.backward()`를 통해 기울기를 계산한 후, `optimizer.step()`으로 수정됨
   - 오차가 줄어들면서 전역 최소점(global minimum)으로 가기 위해 W와 b를 계속 수정

3. **기울기 누적 방지**:
   - 순환 신경망(RNN)에서는 기울기 누적이 필요하지만, 그 외의 모델에서는 `optimizer.zero_grad()`로 기울기를 초기화하여 누적을 방지해야 함

    # 파라미터 업데이트
    optimizer.step()
